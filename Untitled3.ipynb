```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
X = [1,   1.5, 2,   3, 5, 6]
Y = [1.8, 1,   3.6, 2, 1, 3]

plt.plot(X, Y) 
plt.show()
```

```python
plt.scatter(X,Y)  # рисуем точки
plt.title("Просто точечки")
plt.xlabel("Ось X") #x и y заданы в прошлом окне
plt.ylabel("Ось Y")
plt.show()
```

```python
plt.scatter(X, Y)
plt.xlim(0,4) #Ограничили окно вывода для 2х осей
plt.ylim(0,2)
plt.show()
```

```python
plt.scatter(X, Y, s=[50], c='magenta', marker='p')
plt.show()
```

```python
markers = ['o', '*', '8', 'D', '^', 'P']

for x, y, m in zip(X, Y, markers):
    plt.scatter([x],[y], marker=m, s=100)
plt.show()
```

```python
dots = ['Первая', "Вторая", "Третья", "Четвертая", "Пятая", "Шестая"]

for x, y, m, d in zip(X, Y, markers, dots):
    plt.scatter(x, y, marker=m, s=100)
    plt.text(x+0.2, y, d) # +0.2 - это чтобы текст не наползал на маркер, а отрисовывался чуть правее по х и выше по у 
plt.show()
```

```python
plt.plot(range(10), linestyle='-.', marker='*', c='yellow', markevery=[5, 6,7])
plt.show()
```

```python
# частоты слов по текстам

A = [2, 8, 8, 4, 10, 9, 7, 9, 7, 5]
B = [3, 6, 9, 9, 9, 10, 0, 3, 5, 7]

plt.plot(range(10), A, c='g', label="cats")
plt.plot(range(10), B, c='r', label="dogs")
plt.title("Котики и пёсики")
plt.xlabel("Номер текста")
plt.ylabel("Абсолютная частота")
plt.legend()

plt.show()
```

```python
import os
with open("/Users/mac/Desktop/Snake/DH-Homework/Alice.txt", "r", encoding= "utf-8") as a:
    text = a.read ()
    print (text)
```

```python
words = []
for word in text.split():
    word = word.strip("’,!?.,!”‘-–""-«»—’*::;‘“’")
    if len(word) != 0:
        words.append(word)
```

```python
smallalice =[w.lower() for w in words]
```

```python
smallalica = ' '.join(smallalice)
```

```python
from nltk.tokenize import word_tokenize

print(word_tokenize(smallalica))
```

```python
import nltk
```

```python
b = [b for b in nltk.bigrams(word_tokenize(smallalica))]
```

```python
a = [b for b in nltk.trigrams(word_tokenize(smallalica))]
```

```python
from collections import Counter
bigram_counts = Counter (b)
print (bigram_counts.most_common (20))
```

```python
from collections import Counter
trigram_counts = Counter (a)
print (trigram_counts.most_common (20))
```

```python
Aliseandwords = nltk.pos_tag(word_tokenize(smallalica))
```
